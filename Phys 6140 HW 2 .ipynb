six sided dice, how many rolls to get one of each value

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from numpy import random
```

```python
runs = 1000

Dice = [1,2,3,4,5,6]
k = [1,1,1,1,1,1]

rolls = []
totalrolls = []
totalcounts = 0

for j in range(runs):
    k = [1,1,1,1,1,1]
    count = 0
    i = 0
    while i < 6:
        pick = random.choice(Dice)
        
        if pick == 6 and k[0] == 1:
            k[0] = 0
            i = i + 1
        if pick == 1 and k[1] == 1:
            k[1] = 0
            i = i + 1
        if pick == 2 and k[2] == 1:
            k[2] = 0
            i = i + 1
        if pick == 3 and k[3] == 1:
            k[3] = 0
            i = i + 1
        if pick == 4 and k[4] == 1:
            k[4] = 0
            i = i + 1
        if pick == 5 and k[5] == 1:
            k[5] = 0
            i = i + 1
        count = count + 1
    totalcounts = totalcounts + count
    totalrolls.append(totalcounts/(j+1))
    rolls.append(count)
    
AvgRoll = sum(rolls)/runs
PlotRuns=[k for k in range(runs)]
print(AvgRoll)
```

```python
plt.plot(PlotRuns, totalrolls,'r')
plt.axhline(y = AvgRoll, color='black', linestyle='--')
plt.ylabel('number of trials')
plt.xlabel('fraction of the sample')
plt.title(' Smoking population fraction with replacment')
plt.xscale('log')
plt.show()
```

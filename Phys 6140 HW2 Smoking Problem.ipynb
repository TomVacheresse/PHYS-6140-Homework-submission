```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from numpy import random
```

```python
runs = 100000

Select = 10
Population = 1000
NumSmoke = 350
PopList = [*[0 for k in range(Population-NumSmoke)],*[ 1 for k in range(NumSmoke)]]

CountSmoke = 0
CountSmokeTot = 0
Countnull = 0
SmokePlot =[]

for i in range(runs):
    CountSmoke = 0
    for j in range(Select):
            pick = random.choice(PopList)
            CountSmoke = CountSmoke + pick
    if CountSmoke == 4:
        CountSmokeTot = CountSmokeTot + 1
        SmokePlot.append(CountSmokeTot/(i+1))

print(CountSmokeTot/runs)

PlotRuns=[k for k in range(CountSmokeTot)]
```

```python
plt.plot(PlotRuns, SmokePlot,'r')
plt.axhline(y= CountSmokeTot/(runs), color='black', linestyle='--')
plt.ylabel('number of trials')
plt.xlabel('fraction of the sample')
plt.title(' Smoking population fraction with replacment')
plt.xscale('log')
plt.show()
```
